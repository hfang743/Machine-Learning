```python
import numpy as np
import pandas as pd
from sklearn.datasets.california_housing import fetch_california_housing
import matplotlib.pyplot as plt
cal_housing = fetch_california_housing('.')
```

```python
from bokeh.plotting import figure, show
from bokeh.layouts import layout, widgetbox
from bokeh.models import ColumnDataSource, HoverTool, BoxZoomTool, ResetTool, PanTool
from bokeh.models.widgets import Slider, Select, TextInput, Div
from bokeh.models import WheelZoomTool, SaveTool, LassoSelectTool
from bokeh.io import curdoc
from functools import lru_cache
```

```python
print(cal_housing.DESCR)
```

```python
X = cal_housing.data
avg_price = cal_housing.target
feature_names = cal_housing.feature_names
feature_names
```

```python
df=pd.DataFrame(np.hstack([X,avg_price[:,np.newaxis]]),columns=np.hstack([feature_names,'price']))
```

```python
df.head()
```

```python
plt.scatter(df.Longitude,df.Latitude,alpha=0.5,c=df.price,s=df.Population/50,label='population')
plt.colorbar()
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
